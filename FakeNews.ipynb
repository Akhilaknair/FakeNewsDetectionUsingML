```python
# def wordopt(text):
#     text=text.lower()
#     text=re.sub(r'\[.*?\]','',text)
#     text=re.sub(r'\\W','',text)#replaces any non-word character (anything that isn't a letter, number, or underscore) with a space.
#     text=re.sub(r'https?://\S+|www\.\S+','',text)# \S is a shorthand character class in regular expressions that matches any non-whitespace character
#     text=re.sub(r'<.*?>+','',text)
#     text=re.sub(r'[%s]' % re.escape(string.punctuation),'',text)
#     text=re.sub(r'\n','',text)
#     text=re.sub(r'\w*\d\w*','',text)
#     return text


def wordopt(text):
    # Convert to lowercase
    text = text.lower()
    
    # Remove text inside square brackets (if necessary, keep as is)
    text = re.sub(r'\[.*?\]', '', text)
    
    # Remove URLs (this is fine)
    text = re.sub(r'https?://\S+|www\.\S+', '', text)
    
    # Remove HTML tags (if needed, but can be left out if text isn't HTML)
    text = re.sub(r'<.*?>+', '', text)
    
    # Preserve punctuation by removing only unwanted symbols, not all punctuation
    # Remove unwanted characters but keep useful punctuation like periods and commas
    text = re.sub(r'[^\w\s.,!?\'"-]', '', text)
    
    # Remove newlines
    text = re.sub(r'\n', '', text)

    return text
    
```

```python
data['text']=data['text'].apply(wordopt)
```

```python
#defining dep and indep variable
x=data['text']
y=data['class']
```

```python
x_train,x_test,y_train,y_test=train_test_split(x,y,test_size=0.25)
```

```python
from sklearn.feature_extraction.text import TfidfVectorizer #commonly used in natural language processing (NLP) and text mining
vectorization=TfidfVectorizer()
xv_train=vectorization.fit_transform(x_train)
xv_test=vectorization.transform(x_test)
```

```python
#logistic regression
from sklearn.linear_model import LogisticRegression
LR=LogisticRegression()
LR.fit(xv_train,y_train)
```

```python
pred_lr=LR.predict(xv_test)
LR.score(xv_test,y_test)
```

```python
print(classification_report(y_test,pred_lr))
```

```python
# decisiontree classifier
from sklearn.tree import DecisionTreeClassifier
DT=DecisionTreeClassifier()
DT.fit(xv_train,y_train)
```

```python
pred_dt=DT.predict(xv_test)
DT.score(xv_test,y_test)
```

```python
print(classification_report(y_test,pred_dt))
```

```python
from sklearn.ensemble import GradientBoostingClassifier
GB=GradientBoostingClassifier(random_state=0)
GB.fit(xv_train,y_train)
```

```python
pred_gb=GB.predict(xv_test)
GB.score(xv_test,y_test)
```

```python
print(classification_report(y_test,pred_gb))
```

```python
from sklearn.ensemble import RandomForestClassifier
RF=RandomForestClassifier(random_state=0)
RF.fit(xv_train,y_train)
```

```python
pred_rf=RF.predict(xv_test)
RF.score(xv_test,y_test)
```

```python
print(classification_report(y_test,pred_rf))
```

```python
#model testing and  a label funcn
def output_label(n):
    if n == 0:
        return "Fake News"
    elif n == 1:
        return "Not a Fake News"



        

def manual_testing(news):
    testing_news={"text":[news]} #create dic
    new_def_test=pd.DataFrame(testing_news) #dic to dataframe
    new_def_test["text"]=new_def_test["text"].apply(wordopt) #Preprocessing the Text
    new_x_test = new_def_test["text"]
    new_xv_test=vectorization.transform(new_x_test)
    pred_LR=LR.predict(new_xv_test)
    pred_DT=DT.predict(new_xv_test)
    pred_GB=GB.predict(new_xv_test)
    pred_RF=RF.predict(new_xv_test)
    
    return print("\n\n LR Prediction:{} \n DT Prediction:{} \n GBC Prediction:{} \n RFC Prediction:{}".format(output_label(pred_LR),output_label(pred_DT[0]),output_label(pred_GB[0]),output_label(pred_RF[0])))
```
